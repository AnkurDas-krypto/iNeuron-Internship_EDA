```python
import pandas as pd
```

```python
data = pd.read_json(r'C:\Users\ankur\OneDrive\Desktop\Internship\food.json')
```

```python
data.head()
```

```python
for i in range(len(data.ingredients)):
    lst = []
    ing = data.ingredients[i]
    for ingred in ing:
        ingredients_text = ingred['text']
        lst.append(ingredients_text)
        l = ' '.join(lst)
        data.ingredients[i] = l
        
    
```

```python
for i in range(len(data.instructions)):
    instructions = []
    ins = data.instructions[i]
    for instruc in ins:
        instructions_text = instruc['text']
        instructions.append(instructions_text)
        instructions_single_string = ' '.join(instructions)
        data.instructions[i] = instructions_single_string
```

```python
data.head()
```

```python
non_veg = ['turkey', 'chicken', 'egg', 'fish', 'milk', 'beef', 'pork', 'salmon']
```

```python
data['title'] = data.title.apply(lambda x: str(x.lower()))
```

```python
data.head()
```

```python
import numpy as np    
data['type'] = np.nan    #creating column 'type'
```

```python
data.head()
```

```python
lst = []               
for i in data.title:
    lst.append(i)         #putting all title values into the list 'lst'
    
    

#for nonveg ---> 1   for veg ---> 0

for i in range(len(lst)):
    count = 0
    j = lst[i].split(" ")   #getting words for each list values
    for k in j:
        if  k in non_veg:
            data['type'][i] = 1
            break
        else:
            count = count+1      #traversing through j into the last word, if last word is also not in 'non_veg list' then 0
            if(count == len(j)):
                data['type'][i] = 0
                break
            continue
```

```python
data['type'] = data.type.astype(int)    #converting float to int 
```
